```python
import pandas as pd

df = pd.read_csv('../data/hcpcsevents.csv')
```

```python
filtered = df[df["hcpcs_cd"].str.match('^[A-Z]')]
filtered_final = filtered
filtered_final
```

```python
previous_row = ''
counter = 1
for index,row in filtered_final.iterrows():
    if row['hadm_id'] == previous_row:
        counter = counter + 1
    else:
        counter = 1
    filtered_final.at[index,'seq_num']= counter
    previous_row = row['hadm_id']

filtered_final.to_csv('../data/hcpcs_filtered')
```
